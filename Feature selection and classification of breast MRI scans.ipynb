```python
# Izdvajanje slajseva koji imaju lezije samo iz foldera koji sadrže slike PRE
# dodavanja kontrasta i prva 3 prolaza NAKON dodavanja kontrasta
# (folderi: pre, post_1, post_2, post_3)

import os
import numpy as np
from skimage import io

load_root = "D:\FTN\Master\Big data - upravljanje i analiza\Projekat\Baza\manifest-1693924493431\\" # početni deo putanje sa koje ce se učitavati slajsevi
save_root = "D:\FTN\Master\Big data - upravljanje i analiza\Projekat\Baza\izdvojeni_slajsevi\\"     # početni deo putanje na kojoj ce se čuvati izdvojeni slajsevi

sub_folders = ["DICOM_Images", "lesion_masks", "non_lesion_masks"] # folderi koji će čuvati slike sa slajsevima koji sarže lezije, maske koje određuju gde se lezije nalaze i njima simetrične maske

for i in range(csv_data.shape[0]):
    original_path = csv_data.loc[i, "original_path_and_filename"].replace("/", "\\")

    p = original_path.split("\\")[2] # izdvajanje samo polja koje sadrži broj prolaza/serije
    if p == "post_1" or p == "post_2" or p == "post_3" or p == "pre":
        current_slice = int(original_path[-7:].split(".")[0]) # broj trenutnog slajsa na osnovu putanje

        patient_num = int(original_path.split("\\")[1][-3:]) # izdvajanje rednog broja pacijenta na osnovu putanje

        # Smeštanje vrednosti prvog i poslednjeg slajsa na kojem je lezija vidljiva u promenljive
        s = excel_data.loc[patient_num-1, "Start Slice"]
        e = excel_data.loc[patient_num-1, "End Slice"]

        # Smeštanje koordinata lezija u promenljive
        sr = excel_data.loc[patient_num-1, "Start Row"]
        er = excel_data.loc[patient_num-1, "End Row"]
        sc = excel_data.loc[patient_num-1, "Start Column"]
        ec = excel_data.loc[patient_num-1, "End Column"]

        if current_slice > s and current_slice < e:

            # 1. Učitavanje slika sa klasične putanje
            classic_path = csv_data.loc[i, "classic_path"].replace("/", "\\")
            load_path = load_root + classic_path
            ds = dicom.dcmread(load_path)

            for sub_f in sub_folders: # sub_f je promenljiva koja može biti: DICOM_Images, lesion_masks, non_lesion_masks

                # 2. Pravljenje foldera gde će se čuvati slajsevi/maske
                original_path = csv_data.loc[i, "original_path_and_filename"].replace("/", "\\")
                original_path = original_path.replace("DICOM_Images", sub_f)
                new_folder_path = os.path.split(save_root + original_path)[0] # putanja do novih foldera
                #print(new_folder_path)
                if not os.path.exists(new_folder_path):
                   os.makedirs(new_folder_path)                               # pravljenje foldera na putanji

                # 3. Pravljenje putanje na kojoj će se čuvati slajs/maska
                png_path = original_path[:-4] + ".png" # promena dcm ekstenzije na png
                new_path = save_root + png_path        # putanja na kojoj će se čuvati slajs/maska
                #print(new_path)

                # 4. Kreiranje slike
                if sub_f == "DICOM_Images": # ako je u pitanju slika slajsa sa lezijom, smesti je u promenljivu image
                    image = ds.pixel_array
                else:
                    image = np.zeros(ds.pixel_array.shape) # ako nije u pitanju slika sa lezijom, napravi masku dimenzija te slike?
                    image[sr:er, sc:ec] = 255
                    image = image.astype('uint8')
                    if sub_f == "non_lesion_masks":
                        image = np.flip(image, axis=1) # napravi simetričnu masku

                # 5. Čuvanje dobijene slike
                io.imsave(new_path, image, check_contrast=False)
```

```python
# Izvlačenje radiomics obeležja

import os # neophodno za navigaciju do ulaznih podataka
import glob
from radiomics import featureextractor # ovaj modul se koristi za interakciju sa pyradiomics-om
import six

root = "D:\FTN\Master\Big data - upravljanje i analiza\Projekat\Baza\izdvojeni_slajsevi\\DICOM_Images\\"

df = pd.DataFrame()

params = {}
params['label'] = 255
params['normalize'] = True  # normalizacija na nivou slike

extractor = featureextractor.RadiomicsFeatureExtractor(**params)

# odabir grupa obeležja koje želimo da budu izračunata
extractor.disableAllFeatures()
extractor.enableFeatureClassByName('firstorder')
extractor.enableFeatureClassByName('glcm')
extractor.enableFeatureClassByName('glszm')
extractor.enableFeatureClassByName('glrlm')


patient_list = glob.glob(f'{root}/*')
for p_path in patient_list:
    patient_num = int(p_path.split("\\")[-1][-3:])
    bilateral = clinical_features.loc[patient_num, "Bilateral Information"] # informacija o postojanju lezije na obe dojke (0 = nema lezije na obe dojke, 1 = postoji lezija na obe dojke)

    # Smeštanje vrednosti prvog i poslednjeg slajsa na kojem je lezija vidljiva u promenljive
    s = excel_data.loc[patient_num-1, "Start Slice"]
    e = excel_data.loc[patient_num-1, "End Slice"]
    central_slice = (s + e)//2              #izračunavanje centralnog slajsa za tu seriju

    series_list = glob.glob(f'{p_path}/*/') # pravi listu putanja do foldera koji sadrži serije
    for sr_path in series_list:
        series_name = sr_path.split("\\")[-2] # ime serije (može biti pre, post_1, post_2 ili post_3)
        if series_name == "pre":
            series_num = 0
        else:
            series_num = int(series_name[-1]) # uzima poslednji karakter imena foldera i pretvara ga u broj

        slices_list = glob.glob(f'{sr_path}/*') # lista koja sadrži putanje do konkretnih slika slajseva

        for sl_path in slices_list:             # posmatramo svaku sliku
            if int(sl_path[-7:].split(".")[0]) == central_slice: # radiomics obeležja se računaju samo da centralni slajs
                lesion_mask_path = sl_path.replace("DICOM_Images", "lesion_masks") # promena putanje da bi odgovarala maskama lezija
                non_lesion_mask_path = sl_path.replace("DICOM_Images", "non_lesion_masks") # promena putanje da bi odgovarala maskama bez lezija

                result_lesion = extractor.execute(sl_path, lesion_mask_path) # računa radiomics obeležja za slučaj kada je lezija obuhvaćena maskom

                central_slice_col = {"CentralSlice" : central_slice}
                series_col = {"Series" : series_num}
                patient_col = {"PatientNumber" : patient_num}

                label_col = {"Label" : 1} # kad u petlji naiđe osnovna maska, pridružujemo labelu 1

                # uvek tražimo obeležja za osnovnu masku
                res_lesion = {**label_col, **result_lesion}
                res_lesion = {**central_slice_col, **res_lesion}
                res_lesion = {**series_col, **res_lesion}
                res_lesion = {**patient_col, **res_lesion}

                df = df.append(res_lesion, ignore_index=True)

                if bilateral == 0:        # samo za slučaj kad ne postoji bilateralni kancer, tražimo obeležja za simetričnu masku

                    result_non_lesion = extractor.execute(sl_path, non_lesion_mask_path)

                    label_col_n = {"Label" : 0} # kad u petlji naiđe simetrična maska pridružujemo labelu 0

                    res_non_lesion = {**label_col_n, **result_non_lesion}
                    res_non_lesion = {**central_slice_col, **res_non_lesion}
                    res_non_lesion = {**series_col, **res_non_lesion}
                    res_non_lesion = {**patient_col, **res_non_lesion}

                    df = df.append(res_non_lesion, ignore_index=True)
```

```python
# kNN klasifikacija
import matplotlib.pyplot as plt
import sklearn
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, ConfusionMatrixDisplay
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.model_selection import GridSearchCV
import numpy as np

# Pravljenje listi koje sadrže trening, test i validacione podatke u sve 4 kombinacije
X_train_list = [X1_train, X2_train, X3_train, X4_train]
X_val_list = [X1_val, X2_val, X3_val, X4_val]
X_test_list = [X1_test, X2_test, X3_test, X4_test]

y_train_list = [y1_train, y2_train, y3_train, y4_train]
y_val_list = [y1_val, y2_val, y3_val, y4_val]
y_test_list = [y1_test, y2_test, y3_test, y4_test]


# trazenje najboljih parametara za model
parameters = {"n_neighbors" : [1, 2, 3, 5, 10], "metric" : ("euclidean", "chebyshev", "manhattan", "hamming")}

for X, y in zip(X_train_list, y_train_list):

  clf = KNeighborsClassifier()
  clf_gs = GridSearchCV(clf, parameters)

  clf_gs.fit(X, y)

  print(clf_gs.best_score_)
  print(clf_gs.best_params_)

conf_mat_list = []
acc_list = []

# kNN klasifikacija za sve 4 kombinacije podataka
for X, y, X_test, y_test in zip(X_train_list, y_train_list, X_test_list, y_test_list):

  classifier = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
  classifier.fit(X, y)

  y_pred = classifier.predict(X_test)

  # Pravljenje matrica konfuzije
  conf_mat = confusion_matrix(y_test, y_pred, labels=classifier.classes_)
  conf_mat_list.append(conf_mat)

  # Računanje tačnosti
  acc = accuracy_score(y_test, y_pred)
  acc_list.append(acc)
```

```python
# Iscrtavanje matrica konfuzije za sve 4 kombinacije podataka
for conf_mat, acc in zip(conf_mat_list, acc_list):

  disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat, display_labels=classifier.classes_)
  disp.plot()
  plt.show()

  print('Accuracy is', round(acc * 100, 2), '%')
```

```python
# Pravljenje dataframe-a u kom su sva obeležja spojena u jedan vektor

vec_asmbl = VectorAssembler(inputCols=df_train_list[0].columns[1:], outputCol="features") # df_train.columns[5:]

# ekstenzija v predstavlja liste koje sadrže dataframe-ove sa vektorizovanim obeležjima
df_train_list_v = []
df_val_list_v = []
df_test_list_v = []

for df_train, df_val, df_test in zip(df_train_list, df_val_list, df_test_list):
  df_train_v = vec_asmbl.transform(df_train).select("features", "Label")
  df_val_v = vec_asmbl.transform(df_val).select("features", "Label")
  df_test_v = vec_asmbl.transform(df_test).select("features", "Label")

  df_train_list_v.append(df_train_v)
  df_val_list_v.append(df_val_v)
  df_test_list_v.append(df_test_v)

df_train_list_v[1].show(5)
```

```python
# Selektor obeležja - bira obeležja koja najviše doprinose klasifikaciji

from pyspark.ml.feature import VarianceThresholdSelector
from pyspark.ml.feature import ChiSqSelector

selector = VarianceThresholdSelector(varianceThreshold=0.6, outputCol="selectedFeatures")

# ekstenzija s predstavlja liste koje sadrže dataframe-ove sa selektovanim obeležjima
df_train_list_s = []
df_val_list_s = []

for df_train_v, df_val_v in zip(df_train_list_v, df_val_list_v):

  sel_model = selector.fit(df_train_v)
  df_train_s = sel_model.transform(df_train_v)
  sel_model = selector.fit(df_val_v)
  df_val_s = sel_model.transform(df_val_v)

  df_train_list_s.append(df_train_s)
  df_val_list_s.append(df_val_s)

print("Features selected by VarianceThresholdSelector:")
df_train_list_s[1].show(5)
```

```python
# Izdvajanje selektovanih obelezja i labela iz dataframe-a i preimenovanje kolone "selectedFeatures" u "features"

# ekstenzija f predstavlja liste koje sadrže finalne dataframe-ove
df_train_list_f = []
df_val_list_f = []

for df_train, df_val in zip(df_train_list_s, df_val_list_s):

  df_train_f = df_train.select("selectedFeatures", "Label").withColumnRenamed('selectedFeatures', 'features')
  df_val_f = df_val.select("selectedFeatures", "Label").withColumnRenamed('selectedFeatures', 'features')

  df_train_list_f.append(df_train_f)
  df_val_list_f.append(df_val_f)


df_train_list_f[1].show(5)
```

```python
# Definisanje evaluacije rezultata pomoću AUROC
evaluator = BinaryClassificationEvaluator(labelCol='Label', metricName='areaUnderROC')
```

```python
# RandomForest klasifikacija
import matplotlib.pyplot as plt

rf = RandomForestClassifier(labelCol='Label', numTrees=200, maxDepth=5)

models = []
accs = []
aurocs = []

for df_train, df_val in zip(df_train_list_f, df_val_list_f):

  model = rf.fit(df_train)
  models.append(model)
  pred = model.transform(df_val)

  auroc = evaluator.evaluate(pred)

  trainingSummary = model.summary
  roc = trainingSummary.roc.toPandas()

  plt.plot(roc['FPR'], roc['TPR'])
  plt.ylabel('False Positive Rate')
  plt.xlabel('True Positive Rate')
  plt.title('ROC Curve')
  plt.show()

  print('AUROC is', round(auroc * 100, 2), '%')
  aurocs.append(auroc)

  acc = trainingSummary.accuracy
  print('Accuracy is', round(acc * 100, 2), '%')
  print("\n")
  accs.append(acc)
```
